#### 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(x=df['Selling Price'],y=df['Rating'])   # for bivariate, price_deatils and tip stars
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
sns.countplot(df, x= "Brand")        # plot the brand name of mobile phones 
plt.xticks(rotation=45, ha="right")
plt.show()
pd.value_counts(df['Brand'])         # determine the numerical value of the name
```

```python
# for multivariate - Selling Price, Rating and Brand, hue is used here

sns.scatterplot(x=df['Selling Price'],y=df['Rating'],hue=df['Brand'])

# The following scatter plot helps us understand how price and ratings are distributed across various brands
```

### 2. Bar Plot (Numerical - Categorical)

```python
# what is average price of a mobile phone
# x-axis = categorical and y-axis = numerical


plt.figure(figsize=(12, 6))
sns.barplot(x=df['Brand'], y=df['Selling Price'])
plt.xticks(rotation=45, ha="right")
plt.title('Selling Price by Brand')
plt.show()
```

### 3. Box Plot (Numerical - Categorical)

```python
# boxplot is usually used to identify the outliers
plt.figure(figsize=(12, 6))

# Create the boxplot for 'Brand' vs 'Rating'
sns.boxplot(x='Brand', y='Rating', data=df)
plt.xticks(rotation=45, ha="right")
plt.title('Boxplot of Ratings by Brand')
plt.show()
```

```python
# Boxplot: Impact of Brand and Memory on Selling Price

plt.figure(figsize=(12, 6))
sns.boxplot(x='Brand', y='Selling Price', hue='Memory', data=df)
plt.xticks(rotation=45, ha='right')
plt.title('Selling Price of Mobiles by Brand and Memory Size')
plt.show()
```

### 4. Distplot (Numerical - Categorical)

```python
# Creatinging a binary column for ratings
df['High Rating'] = df['Rating'].apply(lambda x: 1 if x > 4 else 0)
```

```python
plt.figure(figsize=(12, 6))

# low ratings
sns.kdeplot(df[df['High Rating'] == 0]['Rating'], label='Low Rating (<= 4)', color='blue', fill=False)

# high ratings
sns.kdeplot(df[df['High Rating'] == 1]['Rating'], label='High Rating (> 4)', color='orange', fill=False)

plt.title('Distribution of Ratings by High and Low Ratings')
plt.xlabel('Rating')
plt.ylabel('Density')
plt.legend()

plt.show()
```

### 5. HeatMap (Categorical - Categorical)

```python
# Creating a binary column for high ratings
df['High Rating'] = df['Rating'].apply(lambda x: 1 if x > 4 else 0)
```

```python
rating_crosstab = pd.crosstab(df['Brand'], df['High Rating'], rownames=['Brand'], colnames=['High Rating'])
print(rating_crosstab)

# 1 indicates a rating greater than 4 and 0 indicates 4 or less
```

```python
df['High Rating'] = df['Rating'].apply(lambda x: 1 if x > 4 else 0)
```

```python
crosstab = pd.crosstab(df['Brand'], df['High Rating'])
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(crosstab, annot=True, cmap='YlGnBu', fmt='g')
plt.title('Heatmap')
plt.xlabel('High Rating (1 = > 4, 0 = <= 4)')
plt.ylabel('Brand')

plt.show()
```

```python
# the percentage or numbers of the above graph or Rating of each mobile brand in percentage
max_rating = 5

percentage_ratings = (df.groupby('Brand')['Rating'].mean() / max_rating) * 100

print(percentage_ratings)
```

```python
(df.groupby('Brand')['Rating'].mean() * 100).plot(kind='bar')
```
